```python
import pandas as pd

df_survey = pd.read_csv('survey_results_public.csv', encoding='utf-8', low_memory=False)
df_schema = pd.read_csv('survey_results_schema.csv', encoding='utf-8')
```

## Підрахунок загальної кількості респондентів

```python
# підрахунок унікальних значень у колонці ResponseId, яка є ідентифікатором кожного учасника
unique_respondent = df_survey['ResponseId'].nunique()
unique_respondent
```

## Аналіз повноти відповідей респондентів

```python
# отримуємо список питань з файлу survey_results_schema (поле "qname")
questions = df_schema['qname'].dropna().unique()
# знаходимо перетин питань та реальних колонок у датасеті за допомогою інтерсекції множин
actual_questions = set(questions) & set(df_survey.columns)
# створюємо датафрейм лише з цими питаннями
df_questions = df_survey[list(actual_questions)]
# рахуємо кількість пропусків у кожному рядку
missing_per_row = df_questions.isna().sum(axis=1)
# розраховуємо кількість респондентів із повним набором відповідей
complete_respondents = (missing_per_row == 0).sum()
complete_respondents
```

## Статистичний аналіз досвіду респондентів

```python
# обчислення статистичних показників для стовпчика WorkExp
workexp_stats = pd.Series({
    'Mean': df_survey['WorkExp'].mean().round(1),
    'Median': df_survey['WorkExp'].median(),
    'Mode': df_survey['WorkExp'].mode()[0]
})
print(f'Measures of central tendency:')
workexp_stats
```

## Аналіз віддаленої роботи

## Визначення популярності Python

```python
# розраховуємо кількість респондентів які працювали з Python
python_users = df_survey['LanguageHaveWorkedWith'].str.contains('Python', na=False).sum()
# розраховуємо загальну кількість респондентів
total_resp = df_survey.shape[0]
# розраховуємо процентну частку тих, кто працював з Python
percent_python = python_users / total_resp
print( f'Percent of respondents who worked with Python: {percent_python:.2%}')
```

## Аналіз шляхів навчання програмуванню

```python
# перевіряємо можливі значення навчання в стовпці LearnCode
df_survey['LearnCode'].unique()
```

```python
# визначаємо текст, який відповідає онлайн-курсам
pattern = 'Online Courses or Certification (includes all media types)'
# створюємо новий стовпець, де True - респондент навчався на онлайн-курсах
df_survey['learned_with_online_courses'] = df_survey['LearnCode'].str.contains(pattern, na=False, regex = False)
#розраховуємо кількість респондентів, що навчались на онлайн-курсах
online_courses_count = df_survey['learned_with_online_courses'].sum()
online_courses_count
```

## Географічний аналіз компенсації Python-розробника

```python
# відфільтровуємо респондентів, які працювали з Python
python_users = df_survey[df_survey['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
# обираємо стовпці з якими будемо працювати
python_comp = python_users[['Country', 'ConvertedCompYearly']]
# прибираємо рядки, де відсутнє значення зарплати
python_comp = python_comp.dropna(subset=['ConvertedCompYearly'])
# групуємо дані за країною і рахуємо середню та медіану компенсації
comp_by_country = python_comp.groupby('Country')['ConvertedCompYearly'].agg(
    mean_comp = 'mean',
    median_comp = 'median',
)
# округляємо значення стовпця mean_comp до 1 знака після коми
comp_by_country['mean_comp'] = comp_by_country['mean_comp'].round(1)
# сортуємо за медіанною компенсацією
comp_by_country_sorted = comp_by_country.sort_values(by=['median_comp'], ascending=False)
comp_by_country_sorted
```

## Аналіз освіти найбільш оплачуваних спеціалістів

```python
# обираємо потрібні колонки
education_analysis = df_survey[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]
# прибираємо пропуски у стовпці ConvertedCompYearly
education_analysis = education_analysis.dropna(subset=['ConvertedCompYearly'])
# сортуємо за компенсацією у спадному порядку
education_analysis_sorted = education_analysis.sort_values(by='ConvertedCompYearly', ascending=False)
# перевстановлюємо індекс
education_analysis_sorted = education_analysis_sorted.reset_index(drop=True)
education_analysis_sorted.head(5)
```

## Аналіз популярності Python по віковим категоріям

```python
# перевіряємо зазначені вікові категорії у колонці Age
df_survey['Age'].unique()
```

```python
# відфільтровуємо респондентів, які працювали з Python
df_survey['python_users'] = df_survey['LanguageHaveWorkedWith'].str.contains('Python', na=False)
# групуємо по віковій категорії
age_python_stats = df_survey.groupby('Age').agg(
    total_resp=('ResponseId', 'count'),
    python_users=('python_users', 'sum'),
)
# розраховуємо відсоток користувачів Python
age_python_stats['percent_python'] = (age_python_stats['python_users'] / age_python_stats['total_resp']) * 100
# округлюємо значення стовпця percent_python
age_python_stats['percent_python'] = age_python_stats['percent_python'].apply(lambda x:f'{x:.1f}%')
age_python_stats
```

## Аналіз індустрій серед високооплачуваних віддалених працівників

```python
# знаходимо 75-й перцентиль компенсації
percentile75 = df_survey['ConvertedCompYearly'].quantile(0.75)
# фільтруємо респондентів, які заробляють більше цього значення
high_paid = df_survey[df_survey['ConvertedCompYearly'] > percentile75]
# залишаємо тих, кто працює віддалено
remote_high_paid = high_paid[high_paid['RemoteWork'] == 'Remote']
# визначаємо найпопулярніші індустрії серез цієї групи
industries = remote_high_paid['Industry'].value_counts()
# перетворюємо у DataFrame та додаємо заголовки
industries_df = industries.reset_index()
industries_df.columns = ['Industry', 'Amount']
industries_df
```
